```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('datasets/Customer.csv')
invoice = pd.read_csv('datasets/Invoice.csv')
```

```python
customers.head()
```

```python
invoice.head()
```

```python
target_df = pd.merge(invoice,customers)
```

```python
target = target_df[['CustomerId', 'Total']]
target = target.groupby('CustomerId').agg({'Total':'sum'})
target
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
X_scaled = target.values
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(range(1, 10), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X_scaled)
p=kmeans.predict(X_scaled)
```
